```python
# Import Dependencies
import pandas as pd
```

```python
# Make a reference to the books.csv file path
import os
path = os.path.join("Resources","books.csv")
# Import the books.csv file as a DataFrame with utf-8 encoding.
books_df = pd.read_csv(path,encoding="UTF-8")
```

```python
#  Get the date types of each column.
books_df.dtypes
```

```python
# Get a list of all columns within the DataFrame
books_df.columns
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
# Only keep: "isbn", "original_publication_year", "original_title", "authors",
# "ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"
reduce_df = books_df[{"isbn", "original_publication_year", "original_title", "authors", "ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"}].copy()
reduce_df.head()
```
